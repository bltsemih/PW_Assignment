```python
import geopandas as gpd
```

```python
districts = gpd.read_file("districts.geojson")
buildings =  gpd.read_file("buildings.geojson")
```

```python
new = gpd.sjoin(districts,buildings,how="inner",op="intersects")
new["const"] = 1
new = new.groupby("gid_left").sum()
new = new.sort_values("const")
```

```python
new
```

```python
for e in list(new.index[-1:-4:-1]):
  print(districts.name[districts.gid == e].values[0], new.loc[e].const)
```
